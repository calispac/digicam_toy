```python
import os.path
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
plt.figure(figsize=(8, 4))
for path in [
    'digicamtoy/utils/pulse_SST-1M_pixel_0.dat',
    'digicamtoy/utils/pulse_SST-1M_AfterPreampLowGain.dat',
]:
    x, y = np.loadtxt(path, unpack=True)
    y /= y.max()
    plt.plot(x, y, '.:', label=os.path.basename(path))
plt.grid()
plt.legend(loc='upper right')
plt.xlim(-20, 100)
```
